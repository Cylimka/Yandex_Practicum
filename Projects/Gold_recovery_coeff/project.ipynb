Модель должна предсказать коэффициент восстановления золота из золотосодержащей руды. Используйте данные с параметрами добычи и очистки. 

Модель поможет оптимизировать производство, чтобы не запускать предприятие с убыточными характеристиками.

Вам нужно:

1. Подготовить данные;

3. Построить и обучить модель.

Описание данных:
* Технологический процесс
    * Rougher feed — исходное сырье
    * Rougher additions (или reagent additions) — флотационные реагенты: Xanthate, Sulphate, Depressant
    * Xanthate — ксантогенат (промотер, или активатор флотации);
    * Sulphate — сульфат (на данном производстве сульфид натрия);*
    * Depressant — депрессант (силикат натрия).
    * Rougher process (англ. «грубый процесс») — флотация
    * Rougher tails — отвальные хвосты
    * Float banks — флотационная установка
    * Cleaner process — очистка
    * Rougher Au — черновой концентрат золота
    * Final Au — финальный концентрат золота
* Параметры этапов
    * air amount — объём воздуха
    * fluid levels — уровень жидкости
    * feed size — размер гранул сырья
    * feed rate — скорость подачи
    
Наименование признаков должно быть такое: [этап].[тип_параметра].[название_параметра] \
Пример: rougher.input.feed_ag \
Возможные значения для блока [этап]: \
rougher — флотация \
primary_cleaner — первичная очистка \
secondary_cleaner — вторичная очистка \
final — финальные характеристики \
Возможные значения для блока [тип_параметра]: \
input — параметры сырья \
output — параметры продукта \
state — параметры, характеризующие текущее состояние этапа \
calculation — расчётные характеристики \

### Проверка рассчета эффективности

Проверим что эффективность обогащения рассчитана правильно. Для этого напишем функцию по расчету эффективность обогащения.

```python
def efficiency(C, F, T):
    recovercy = ((C * (F - T)) / (F * (C - T))) * 100
    return recovercy
```

Сравним показания нашей функции на обучающей выборке с признаком rougher.output.recovery, найдите MAE между нашими расчётами и значением признака. 

```python
mean_absolute_error(train_df['rougher.output.recovery'], 
                   efficiency(train_df['rougher.output.concentrate_au'],
                             train_df['rougher.input.feed_au'],
                             train_df['rougher.output.tail_au']))
```

MAE слишком маленькое, это значит что разница между нашими вычислениями и изначальными очень маленькое, то есть изначальный расчет эффективность обогащения расчитан правильно

#### Избавимся от выбросов в диапозоне 0-1

```python
def drop_near_zerro(df, material):
    originally_len = len(df)
    count = len(df[df[f'rougher.input.feed_{material}'] < 1])
    df = df.drop(df[df[f'rougher.input.feed_{material}'] < 1].index)
    
    count += len(df[df[f'rougher.output.concentrate_{material}'] < 1])
    df = df.drop(df[df[f'rougher.output.concentrate_{material}'] < 1].index)
    
    count += len(df[df[f'primary_cleaner.output.concentrate_{material}'] < 1])
    df = df.drop(df[df[f'primary_cleaner.output.concentrate_{material}'] < 1].index)
    
    count += len(df[df[f'final.output.concentrate_{material}'] < 1])
    df = df.drop(df[df[f'final.output.concentrate_{material}'] < 1].index)
    
    print(f'Мы потеряли {round(count*100/originally_len,2)}% данных')
    return df
```

```python
for metall in ['au', 'ag', 'pb']:
    print(metall)
    train_df = drop_near_zerro(train_df, metall)
    print()
```

Мы избавились от значений в деапозоне от 0 до 1, для золота, серебра и свинца, потеряв 5.5% данных

## Модель

### Напишем функцию для вычисления sMAPE и итоговой sMAPE

```python
def smape(target, predictions):
    smape = (sum(abs(target-predictions) / ((abs(target) + abs(predictions)) / 2)) / len(target)) * 100
    return smape
```

```python
def total_smape(rougher, final):
    total_sampe = round(abs((pd.Series(0.25 * rougher + 0.75 * final).mean())), 2)
    return total_sampe
```

Создадим кастомный скорер (оценщик) для использования в процессе оценки моделей.

```python
smape_scorer = make_scorer(smape, greater_is_better = False)
```

```python
params = {'max_depth': range(1,10,1),
    'min_samples_split': range(2,10,1),
    'min_samples_leaf': range(2,10,1)
}

model = DecisionTreeRegressor(random_state=RANDOM_STATE)

tree = GridSearchCV(model,
                    params, 
                    scoring = smape_scorer,
                    cv = 5, 
                    n_jobs = -1)

tree_rough = tree.fit(features_train, target_train['rougher.output.recovery'])
tree_final = tree.fit(features_train, target_train['final.output.recovery'])

print('sMAPE модели флотации:', tree_rough.best_score_)
print('sMAPE модели финального этапа:', tree_final.best_score_)
print('Параметры модели для целевых показателей после флоатции', tree_rough.best_params_)
print('Параметры модели для целевых показателей после второй очистки', tree_final.best_params_)
print('Итоговое sMAPE:', total_smape(tree_rough.best_score_, tree_final.best_score_))
```
